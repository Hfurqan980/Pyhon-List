###`Problem 6:` Find list of common unique items from two list. and show in increasing order

```
num1 = [23,45,67,78,89,34]
num2 = [34,89,55,56,39,67]
```

`Output:`
```
[34, 67, 89]
```


````python
# Write code here
num1 = [23,45,67,78,89,34]
num2 = [34,89,55,56,39,67]
unique_item = []

for i in num1:
  for j in num2:
    if i == j:
      unique_item.append(i)
print(sorted(unique_item))
````

###`Problem 7:` Sort a list of alphanumeric strings based on product value of numeric character in it. If in any string there is no numeric character take it's product value as 1.

`Input:`
```
['1ac21', '23fg', '456', '098d','1','kls']
```

`Output:`
```
['456', '23fg', '1ac21', '1', 'kls', '098d']

```

````python
from string import digits
# Write code here
def digit_product(s):
    digits = [int(ch) for ch in s if ch.isdigit()]
    if not digits:
        return 1
    product = 1
    for d in digits:
        product *= d
    return product

list1 = ['1ac21', '23fg', '456', '098d', '1', 'kls']

sorted_list = sorted(list1, key=digit_product, reverse=True)
print(sorted_list)


````

### `Problem 10:` Add Space between Potential Words.

**Example:**

Input:

```bash
['campusxIs', 'bestFor', 'dataScientist']
```

Output:
```bash
['campusx Is', 'best For', 'data Scientist']
```

````python
# Code here
list = ['campusxIs', 'bestFor', 'dataScientist']
list1 = []

for w in list:
  word = ""
  for i in w:
    if i.isupper():
      word = word + " "+ i
    else:
      word+=i
  list1.append(word)
print(list1)
````

### `Problem 11:` Write a program that can perform union operation on 2 lists

**Example:**

Input:

```bash
[1,2,3,4,5,1]
[2,3,5,7,8]
```

Output:
```bash
[1,2,3,4,5,7,8]
```

````python
# Write code here
l1 = [1,2,3,4,5,1]
l2 = [2,3,5,7,8]

# one method
l3 = []
for i in l1 + l2:
  if i not in l3:
    l3.append(i)
print("first method = ",l3)

# Second method
# convert the list into set and then apply the union

union = set(l1) | set(l2)
print("second method ",union)
````
